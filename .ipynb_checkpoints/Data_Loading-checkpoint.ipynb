```python
from docx import Document
import re

import pandas as pd
```

```python
docx_file = 'Supp-A.docx'
doc = Document(docx_file)
```

```python
docx2 = 'Supp-B.docx'
doc2 = Document(docx2)
```

```python
pattern = r'>[^\n]+\n([A-Z\n]+)'
```

```python
extracted_text = ''

for paragraph in doc.paragraphs:
    extracted_text += paragraph.text + '\n'
```

```python
matches = re.findall(pattern, extracted_text)
```

```python
protein_sequences_1 = []
protein_sequences_2 = []
for idx, match in enumerate(matches, start=1):
    if idx % 2 == 0:
        protein_sequences_2.append(match.replace('\n', ''))
    else:
        protein_sequences_1.append(match.replace('\n', ''))
```

```python
protein_sequences_1
```

```python
df = pd.DataFrame({
    'protein_sequences_1': protein_sequences_1,
    'protein_sequences_2': protein_sequences_2
})
```

```python
df
```

```python
df.to_csv('positive_protein_sequences.csv', index=False)
```

```python
extracted_text2 = ''

for paragraph in doc2.paragraphs:
    extracted_text2 += paragraph.text + '\n'
```

```python
pattern2 = r'>[^\n]+\n([A-Z\n]+)'
```

```python
matches2 = re.findall(pattern2, extracted_text2)
```

```python
neg_protein_sequences_1 = []
neg_protein_sequences_2 = []
for idx, match in enumerate(matches2, start=1):
    if idx % 2 == 0:
        neg_protein_sequences_2.append(match.replace('\n', ''))
    else:
        neg_protein_sequences_1.append(match.replace('\n', ''))
```

```python
neg_df = pd.DataFrame({
    'neg_protein_sequences_1': neg_protein_sequences_1,
    'neg_protein_sequences_2': neg_protein_sequences_2
})
```

```python
neg_df
```

```python
neg_df.to_csv('negative_protein_sequences.csv', index=False)
```
